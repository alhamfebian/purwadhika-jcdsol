```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import folium
```

```python
tips_data = sns.load_dataset('tips')

tips_data.head()
```

```python
# Bikin histogram pake Mathplotlib

plt.style.use('fivethirtyeight')
plt.hist(
    x = tips_data['total_bill']
)
plt.title('Total Bill Histogram', size=15)
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.legend(['Total Bill from Customer'], loc = 0)
plt.show()
```

```python
# Bikin histogram pake Seaborn -> sebenarnya bikin histogram pake Mathplotlib

sns.histplot(
    data = tips_data,
    x = 'total_bill',
    kde=True,
    legend=True
)

plt.show()
```

```python
# Bikin histogram pake Pandas -> sebenarnya bikin histogram pake Mathplotlib

tips_data['total_bill'].hist()
```

```python
# Bikin Boxplot pake Mathplotlib

plt.boxplot(
    x = tips_data['total_bill'],
    vert = False
)
plt.title('Total Bill Histogram', size=15)
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bikin Boxplot pake Seaborn

sns.boxplot(
    data = tips_data,
    x = 'total_bill'
)
```

```python
# Bikin Multiple Boxplot pake Seaborn 

sns.boxplot(
    data = tips_data,
    x = 'total_bill',
    y = 'day',
    hue = 'day'
)
```

```python
# Bikin Multiple Boxplot pake Pandas

tips_data.boxplot(
    column = 'total_bill',
    rot = 34
)
```

```python
flights_data = sns.load_dataset('flights')
```

```python
# Bikin Lineplot pake Mathplotlip
# Sangat berguna untuk menampilkan data yang ada timeframenya
# Contohnya kayak grafik stock market analysis

plt.plot(
    flights_data[flights_data['month'] == 'Jan']['year'],
    flights_data[flights_data['month'] == 'Jan']['passengers']
)
plt.title('Number of passengers in January, every year')
plt.xlabel('year')
plt.ylabel('Passenger Count')
plt.show()
```

```python
# Bikin Lineplot pake Mathplotlip
# Dengan multiple data

x1 = flights_data[flights_data['month'] == 'Jan']['year']
y1 = flights_data[flights_data['month'] == 'Jan']['passengers']
x2 = flights_data[flights_data['month'] == 'Dec']['year']
y2 = flights_data[flights_data['month'] == 'Dec']['passengers']

plt.plot(
    x1,y1, 'red',
    x2,y2, 'blue',
)
plt.title('Number of passengers January vs December, every year')
plt.xlabel('year')
plt.ylabel('Passenger Count')
plt.show()
```

```python
# Bikin lineplot dengan seaborn

plt.figure(figsize = (16,9))

sns.lineplot(
    data = flights_data,
    x = 'year',
    y = 'passengers',
    hue = 'month'
)
```

```python
# Bikin lineplot dengan Pandas
# Harus bikin data yang sudah terfilter
# Pada contoth ini: filter data flight untuk bulan January

flights_data_january = flights_data[flights_data['month'] == 'Jan']

flights_data_january.plot.line(
    x = 'year',
    y = 'passengers'
)
```

```python
# Tapi kalau kita pake data yang tidak di filter, akan aneh bentuk grafiknya

flights_data.plot.line(
    x = 'year',
    y = 'passengers'
)

plt.title('Number of passengers in January', size = 20)
```

```python
# Bikin Bar Plot Menggunakan Mathplotlib

tips = sns.load_dataset('tips')

tips_by_day = tips[['tip', 'day']].groupby(['day']).mean()
tips_by_day

x = tips['day'].unique().tolist()
y = tips_by_day['tip']

plt.bar(x,y)
plt.xlabel('days')
plt.ylabel('tips mean')
```

```python
plt.barh(x,y)
plt.xlabel('days')
plt.ylabel('tips')
```

```python
# Bikin Bar Plot Menggunakan Seaborn

sns.barplot(
    data=tips,
    x = 'day',
    y = 'tip',
    hue='day',
    errorbar=None
)
plt.title('Barplot: Average Tips Everyday')
```

```python
sns.barplot(
    data=tips,
    x = 'tip',
    y = 'day',
    hue='day',
    errorbar=None
)
plt.title('Barplot: Average Tips Everyday')
```

```python
# Bikin Bar Plot Menggunakan Pandas

tips = sns.load_dataset('tips')

tips_by_day = tips[['tip', 'day']].groupby(['day']).mean()
tips_by_day.reset_index(inplace=True)

tips_by_day.plot.bar(
    x = 'day',
    y = 'tip'
)
plt.title('Tip by Day')
```

```python
# Bikin Piechart menggunakan Mathplotlib

plt.pie(
    x = tips_by_day['tip'],
    labels = tips_by_day['day'],
    autopct='%.2f'
)
```

```python
# Contoh bikin folium (Map) untuk Jakarta
# locationnya bisa dapet dari openstreetmap.org

map_jakarta = folium.Map(
    location = [-6.2, 106.816666],
    zoom_start = 12,
    style = 'OpenStreetMap'
)

# Tambahin Marker
# Monas
folium.Marker(
    [-6.175392, 106.827153],
    popup = '<i>Monas<i>',
    tooltip = 'Ini Monumen Nasional'
).add_to(map_jakarta)


# Kota Tua Jakarta
folium.Marker(
    [-6.1376, 106.8171],
    popup = 'Kota Tua Jakarta',
    tooltip = 'Mari ke Kota Tua Jakarta',
    icon = folium.Icon(
        color = 'red',
        icon = 'info-sign'
    )
).add_to(map_jakarta)


# Blok M
folium.CircleMarker(
    [-6.2446, 106.8006],
    radius = 30,
    popup = 'Blok M',
    tooltip = 'Pandemi selesai, ayo nongkrong di sini',
    color = '#3186cc',
    fill = True,
    fill_color = '#3186cc'
).add_to(map_jakarta)

map_jakarta
```

```python
# Bikin choloropeth map

df = pd.read_csv('pendatang_DKI_Maret_2020.csv')
df.head()
```

```python
df_kel = df[['kelurahan', 'jumlah']]

pendatang_jakarta = "jakarta_by_kelurahan_geo.geojson"
jkt_location = [-6.2, 106.816666]
jkt_map = folium.Map(
    location = jkt_location, zoom_start=12
)
folium.Choropleth(
    geo_data = pendatang_jakarta,
    data = df_kel,
    columns = ['kelurahan', 'jumlah'],
    key_on = 'feature.properties.name',
    fill_color = 'YlOrRd',
    fill_opacity = 0.7,
    line_opacity = 0.5,
    legend_name = "Transmigrant Rate in Jakarta"
).add_to(jkt_map)
jkt_map
```
