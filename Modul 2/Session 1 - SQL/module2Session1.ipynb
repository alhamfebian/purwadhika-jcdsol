```python
# Cara 2 -> Dengan membuat Function

import mysql.connector
import pandas as pd

mydb = mysql.connector.connect(
    host = 'localhost',
    user = 'root',
    passwd = 'Alhamfbn25',
    database = 'world'
)

mycursor = mydb.cursor()

def sql_df(query):
    mycursor.execute(query)
    myresult = mycursor.fetchall()
    df = pd.DataFrame(myresult, columns = mycursor.column_names)
    return df
```

```python
# Mencoba menjalankan Function Query

sql_df(
    '''
        select * 
        from city 
        limit 5
    '''
)
```

```python
# Coba mengerjakan excercise menggunakan pyhton
# Excercise 1

sql_df(
    '''
        select customer_id,rental_id,amount,payment_date
        from sakila.payment
        limit 10;
    '''
)
```

```python
# Excercise 2

sql_df(
    '''
        select title,release_year,rental_duration
        from sakila.film
        where title like "S%";
    '''
)
```

```python
# Excercise 3

sql_df(
    '''
        select 
	        rental_duration as Durasi_Rental, 
            count(rental_duration) as Banyak_Film_Setiap_Durasi_Rental, 
            round(avg(length), 2) as rata_rata_durasi
        from sakila.film
        group by Durasi_Rental;
    '''
)
```

```python
# Excercise 4

sql_df(
    '''
        select 
            title, 
            length, 
            rating
        from sakila.film
        where length > (select avg(length) from sakila.film)
        order by length ASC
        limit 25;
    '''
)
```
