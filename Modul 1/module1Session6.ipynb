```python
# List
listContoh = ['hello', 1,2,3, True]
print(listContoh)

# Index -1 sama aja kayak index paling terakhir
print(listContoh[-1])

# Cara slicing di list
print(listContoh[3:-1])
print(listContoh[1:3])

# Cara slicing pake step
print(listContoh[2:5:3])

# Cara assign new value ke list
listContoh[0] = 'hai'
print(listContoh)

# Cara nambahin data ke list
# Nambahin data ke list dan di taro di paling akhir
listContoh.append('hai')
print(listContoh)

# Nambahin data ke list dan di taro di paling awal
listContoh.insert(2, 'hola')
print(listContoh)

# Cara delete data di list
# Remove data list yang sesuai dengan parameter yang dikasih
listContoh.remove('hola')
print(listContoh)

# Delete data di list yang paling terakhit
listContoh.pop()

# Delete data di list sesuai index yang dikasih
del listContoh[1]

# Menghapus semua value yang ada di list
listContoh.clear()

# Cara looping di list
listContoh = ['hello', 1,2,3, True]
for item in listContoh:
    print(item)

# Cara mengecek apakah ada datanya di list
if 'hola' in listContoh:
    print('ada')
else:
    print('tidak ada')

# Cara mengcopy list
newList = listContoh.copy()
newList[1] = 'Mantul'
print(listContoh)
print(newList)

# Cara sorting di list
listContohSort = [1,3,56,21]
listContohSort.sort()
print(listContohSort)

# Cara ngebalikin urutan di list
listContohReverse = [1,3,56,21]
listContohReverse.reverse()
print(listContohReverse)
```

```python
# List Excercise: Soal 1 Find Min and Max Value
listExercise = [41, 5, 1, 3, 89, 32]
minValue = listExercise[0]
maxValue = listExercise[0]
for value in listExercise:
    if value > maxValue:
        maxValue = value
    elif value < minValue:
        minValue = value
print(f'Min Value: {minValue}')
print(f'Max Value: {maxValue}')
```

```python
# List Excercise: Soal 2 Sort Ascending
listExercise = [41, 5, 1, 3, 89, 32]
for i in range(len(listExercise)):
    for j in range(len(listExercise)):
        if listExercise[i] < listExercise[j]:
            listExercise[i],listExercise[j] = listExercise[j],listExercise[i]
print(listExercise)
```

```python
# List Excercise: Membuat menu di Toko List 2D
print('Selamat Data Di Toko Buah Aldy')

dataTokoBuah = [
    ['Index', 'Nama Buah', 'Stok', 'Stok Tersisa'],
    ['Apel', 5, 10000],
    ['Jeruk', 7, 15000],
    ['Anggur', 9, 20000]
]

for idx, item in enumerate(dataTokoBuah):
    if idx == 0:
        continue
    dataTokoBuah[idx].insert(0,idx-1)

kantongBelanjaUser = [
    ['Nama', 'Qty', 'Harga Per Buah']
]

listMenuToko = [
    '1. Menampilkan Daftar Buah',
    '2. Menambah Buah',
    '3. Menghapus Buah',
    '4. Membeli Buah',
    '5. Keluar Dari Toko'
]

for item in listMenuToko:
    print(item)

menuYangDipilihUser = int(input('Masukkan angka Menu yang ingin di jalankan: '))

print(f'Menu yang dipilih adalah : {menuYangDipilihUser}')

while menuYangDipilihUser != 5:
    if(menuYangDipilihUser == 1): 
        for item in dataTokoBuah:
            print(' | '.join(map(str, item)))
        menuYangDipilihUser = int(input('Masukkan angka Menu yang ingin di jalankan: '))
        print(f'Menu yang dipilih adalah : {menuYangDipilihUser}')

    elif(menuYangDipilihUser == 2):
        buahYangInginDitambahkan = input('Masukan Nama Buah : ')
        stokBuahYangDitambahkan = int(input('Masukan Stok Buah : '))
        hargaBuahYangDitambahkan = int(input('Masukan Harga Buah : '))
        dataTokoBuah.append([buahYangInginDitambahkan, stokBuahYangDitambahkan, hargaBuahYangDitambahkan])
        for idx, item in enumerate(dataTokoBuah):
            if idx in range(0,len(dataTokoBuah)-1):
                continue
            dataTokoBuah[idx].insert(0,idx-1)
        menuYangDipilihUser = int(input('Masukkan angka Menu yang ingin di jalankan: '))
        print(f'Menu yang dipilih adalah : {menuYangDipilihUser}')

    elif(menuYangDipilihUser == 3):
        buahYangMauDiDelete = int(input('Masuk Index Buah Yang Ingin Di Delete : '))
        
        for idx, item in enumerate(dataTokoBuah):
            if idx == 0:
                continue
            if dataTokoBuah[idx][0] == buahYangMauDiDelete:
                del dataTokoBuah[idx]
        
        for idx, item in enumerate(dataTokoBuah):
            if idx == 0:
                continue
            dataTokoBuah[idx][0] = idx-1
        menuYangDipilihUser = int(input('Masukkan angka Menu yang ingin di jalankan: '))
        print(f'Menu yang dipilih adalah : {menuYangDipilihUser}')

    elif(menuYangDipilihUser == 4):
        nomorBuahYangInginDibeli = int(input('Masukan Nomor Buah Yang Ingin Dibeli : '))
        stokBuahYangInginDibeli = int(input('Berapa Buah Yang Ingin Anda Beli? '))

        for idx, item in enumerate(dataTokoBuah):
            if idx == 0:
                continue
            if dataTokoBuah[idx][0] == nomorBuahYangInginDibeli:
                kantongBelanjaUser.append([dataTokoBuah[idx][1], stokBuahYangInginDibeli, dataTokoBuah[idx][3]])

        for item in kantongBelanjaUser:
            print(' | '.join(map(str, item)))
        
        apakahUserMauBeliLagi = input('Apakah anda ingin membeli buah lagi? (Ya/Tidak) ')
        while apakahUserMauBeliLagi.lower() == 'ya':
            nomorBuahYangInginDibeli = int(input('Masukan Nomor Buah Yang Ingin Dibeli : '))
            stokBuahYangInginDibeli = int(input('Berapa Buah Yang Ingin Anda Beli? '))
            
            for idx, item in enumerate(dataTokoBuah):
                if idx == 0:
                    continue
                if dataTokoBuah[idx][0] == nomorBuahYangInginDibeli:
                    kantongBelanjaUser.append([dataTokoBuah[idx][1], stokBuahYangInginDibeli, dataTokoBuah[idx][3]])

            apakahUserMauBeliLagi = input('Apakah anda ingin membeli buah lagi? (Ya/Tidak) ')

        
        for item in kantongBelanjaUser:
            print(' | '.join(map(str, item)))

        totalBelanjaUser = 0
        print('\nDetail Belanja\n')

        for idx in range(1, len(kantongBelanjaUser)):
            print(f'{kantongBelanjaUser[idx][0]} : {kantongBelanjaUser[idx][1]} x {kantongBelanjaUser[idx][2]} = {kantongBelanjaUser[idx][1] * kantongBelanjaUser[idx][2]}')
            totalBelanjaUser += kantongBelanjaUser[idx][1] * kantongBelanjaUser[idx][2]
        print(f'Total Belanja : Rp{totalBelanjaUser}')
        inputUangUser = int(input('Masuk Jumlah Uang: '))
        print(f'Uang anda sebesar: {inputUangUser}')

        while (inputUangUser - totalBelanjaUser) < 0:
            print(f'Transaksi anda dibatalkan\nuangnya kurang sebesar Rp{abs(inputUangUser - totalBelanjaUser)}')
            inputUangUser = int(input('Masuk Jumlah Uang: '))
        if (inputUangUser - totalBelanjaUser) > 0:
            print(f'Terima Kasih\nUang Kembali anda : Rp{inputUangUser - totalBelanjaUser}')
            break
        else:
            print('Terima Kasih')
            break

print('Terima Kasih Sudah Berbelanja di Toko Buah Aldy!')
```

```python
# Zip
data1 = range(1,6)
data2 = ['a','b','c','d']
data3 = (1,3,5,7,9,11)
data4 = 'mnopqrs'

hasilZip1 = zip(data3, data2)
print(hasilZip1)
# <zip object at 0x0000019410C81600>
listHasilZip1 = list(hasilZip1)
print(listHasilZip1)
# [(1, 'a'), (2, 'b'), (3, 'c'), (4, 'd')]
hasilZip2 = zip(data2,data3,data4)
print(hasilZip2)
# <zip object at 0x0000019410C816C0>
tupleHasilZip2 = tuple(hasilZip2)
print(tupleHasilZip2)
# (('a', 1, 'm'), ('b', 3, 'n'), ('c', 5, 'o'), ('d', 7, 'p'))
hasilZip3 = zip(data4,data1)
print(hasilZip3)
# <zip object at 0x0000019410C81740>
dictHasilZip3 = dict(hasilZip3)
print(dictHasilZip3)
# {'m': 1, 'n': 2, 'o': 3, 'p': 4, 'q': 5}

# Looping di Zip()
# example 1
dataNama = ['Andi','Budi','Ceci','Dedi']
dataUmur = [24,27,35,55]
for nama,umur in zip(dataNama,dataUmur):
    print('Nama : {}\nUmur : {}'.format(nama,umur))
# example 2
for i,j in zip(range(5),range(10,0,-2)):
    print('i = {} & j = {}'.format(i,j))
```

```python
# Dictionary Excercise: Memakai data types dictionary, option 1: List of Dictionary
print('Selamat Data Di Toko Buah Aldy')


dataTokoBuah = [
    {
        'index': 0,
        'namaBuah': 'Apel',
        'stokBuah': 5,
        'hargaBuah': 10000
    },
    {
        'index': 1,
        'namaBuah': 'Jeruk',
        'stokBuah': 7,
        'hargaBuah': 15000
    },
    {
        'index': 2,
        'namaBuah': 'Anggur',
        'stokBuah': 9,
        'hargaBuah': 20000
    }
    
]

kantongBelanjaUser = []


listMenuToko = [
    '1. Menampilkan Daftar Buah',
    '2. Menambah Buah',
    '3. Menghapus Buah',
    '4. Membeli Buah',
    '5. Keluar Dari Toko'
]

for item in listMenuToko:
    print(item)

menuYangDipilihUser = int(input('Masukkan angka Menu yang ingin di jalankan: '))

print(f'Menu yang dipilih adalah : {menuYangDipilihUser}')

column_index = max(len(str(item["index"])) for item in dataTokoBuah) + 2
column_namaBuah = max(len(item["namaBuah"]) for item in dataTokoBuah) + 2
column_stokBuah = max(len(str(item["stokBuah"])) for item in dataTokoBuah) + 2
column_hargaBuah = max(len(str(item["hargaBuah"])) for item in dataTokoBuah) + 2

print(f"{'Index':<{column_index}} {'Nama Buah':<{column_namaBuah}} {'Stok Buah':<{column_stokBuah}} {'Harga Buah':<{column_hargaBuah}}")
print("-" * (column_index + column_namaBuah + column_stokBuah + column_hargaBuah))

while menuYangDipilihUser != 5:
    if(menuYangDipilihUser == 1): 
        for item in dataTokoBuah:
            print(f"{item['index']:<{column_index}} {item['namaBuah']:<{column_namaBuah}} {item['stokBuah']:<{column_stokBuah}} {item['hargaBuah']:<{column_hargaBuah}}")
        menuYangDipilihUser = int(input('Masukkan angka Menu yang ingin di jalankan: '))
        print(f'Menu yang dipilih adalah : {menuYangDipilihUser}')

    elif(menuYangDipilihUser == 2):
        buahYangInginDitambahkan = input('Masukan Nama Buah : ')
        stokBuahYangDitambahkan = int(input('Masukan Stok Buah : '))
        hargaBuahYangDitambahkan = int(input('Masukan Harga Buah : '))
        dataTokoBuah.append({
            'index': 0,
            'namaBuah': buahYangInginDitambahkan,
            'stokBuah': stokBuahYangDitambahkan,
            'hargaBuah': hargaBuahYangDitambahkan
        })
        for idx, item in enumerate(dataTokoBuah):
            if idx in range(0,len(dataTokoBuah)-1):
                continue
            dataTokoBuah[idx]['index'] = idx
        menuYangDipilihUser = int(input('Masukkan angka Menu yang ingin di jalankan: '))
        print(f'Menu yang dipilih adalah : {menuYangDipilihUser}')

    elif(menuYangDipilihUser == 3):
        buahYangMauDiDelete = int(input('Masuk Index Buah Yang Ingin Di Delete : '))
        
        for idx, item in enumerate(dataTokoBuah):
            if dataTokoBuah[idx]['index'] == buahYangMauDiDelete:
                del dataTokoBuah[idx]
        
        for idx, item in enumerate(dataTokoBuah):
            dataTokoBuah[idx]['index'] = idx
        menuYangDipilihUser = int(input('Masukkan angka Menu yang ingin di jalankan: '))
        print(f'Menu yang dipilih adalah : {menuYangDipilihUser}')

    elif(menuYangDipilihUser == 4):
        nomorBuahYangInginDibeli = int(input('Masukan Nomor Buah Yang Ingin Dibeli : '))
        stokBuahYangInginDibeli = int(input('Berapa Buah Yang Ingin Anda Beli? '))

        for idx, item in enumerate(dataTokoBuah):
            if dataTokoBuah[idx]['index'] == nomorBuahYangInginDibeli:
                kantongBelanjaUser.append({
                    'namaBuah': dataTokoBuah[idx]['namaBuah'],
                    'stokBuah': stokBuahYangInginDibeli,
                    'hargaPerBuah': dataTokoBuah[idx]['hargaBuah']
                })

        for item in kantongBelanjaUser:
            print(f"{item['namaBuah']:<{column_namaBuah}} {item['stokBuah']:<{column_stokBuah}} {item['hargaPerBuah']:<{column_hargaBuah}}")
        
        apakahUserMauBeliLagi = input('Apakah anda ingin membeli buah lagi? (Ya/Tidak) ')
        while apakahUserMauBeliLagi.lower() == 'ya':
            nomorBuahYangInginDibeli = int(input('Masukan Nomor Buah Yang Ingin Dibeli : '))
            stokBuahYangInginDibeli = int(input('Berapa Buah Yang Ingin Anda Beli? '))
            
            for idx, item in enumerate(dataTokoBuah):
                if dataTokoBuah[idx]['index'] == nomorBuahYangInginDibeli:
                    kantongBelanjaUser.append({
                        'namaBuah': dataTokoBuah[idx]['namaBuah'],
                        'stokBuah': stokBuahYangInginDibeli,
                        'hargaPerBuah': dataTokoBuah[idx]['hargaBuah']
                })

            apakahUserMauBeliLagi = input('Apakah anda ingin membeli buah lagi? (Ya/Tidak) ')

        
        for item in kantongBelanjaUser:
            print(f"{item['namaBuah']:<{column_namaBuah}} {item['stokBuah']:<{column_stokBuah}} {item['hargaPerBuah']:<{column_hargaBuah}}")

        totalBelanjaUser = 0
        print('\nDetail Belanja\n')

        for idx in range(0, len(kantongBelanjaUser)):
            print(f'{kantongBelanjaUser[idx]['namaBuah']} : {kantongBelanjaUser[idx]['stokBuah']} x {kantongBelanjaUser[idx]['hargaPerBuah']} = {kantongBelanjaUser[idx]['stokBuah'] * kantongBelanjaUser[idx]['hargaPerBuah']}')
            totalBelanjaUser += kantongBelanjaUser[idx]['stokBuah'] * kantongBelanjaUser[idx]['hargaPerBuah']
        print(f'Total Belanja : Rp{totalBelanjaUser}')
        inputUangUser = int(input('Masuk Jumlah Uang: '))
        print(f'Uang anda sebesar: {inputUangUser}')

        while (inputUangUser - totalBelanjaUser) < 0:
            print(f'Transaksi anda dibatalkan\nuangnya kurang sebesar Rp{abs(inputUangUser - totalBelanjaUser)}')
            inputUangUser = int(input('Masuk Jumlah Uang: '))
        if (inputUangUser - totalBelanjaUser) > 0:
            print(f'Terima Kasih\nUang Kembali anda : Rp{inputUangUser - totalBelanjaUser}')
            break
        else:
            print('Terima Kasih')
            break

print('Terima Kasih Sudah Berbelanja di Toko Buah Aldy!')
```

```python
# Dictionary Excercise: Memakai data types dictionary, option 2: Dictionary
print('Selamat Data Di Toko Buah Aldy')


dataTokoBuah = {
    'Apel': {
        'Index': 0,
        'Stok': 3,
        'Harga': 10000
    },
    'Jeruk': {
        'Index': 1,
        'Stok': 5,
        'Harga': 15000
    },
    'Anggur': {
        'Index': 2,
        'Stok': 9,
        'Harga': 20000
    }
}

kantongBelanjaUser = {}


listMenuToko = [
    '1. Menampilkan Daftar Buah',
    '2. Menambah Buah',
    '3. Menghapus Buah',
    '4. Membeli Buah',
    '5. Keluar Dari Toko'
]

for item in listMenuToko:
    print(item)

menuYangDipilihUser = int(input('Masukkan angka Menu yang ingin di jalankan: '))

print(f'Menu yang dipilih adalah : {menuYangDipilihUser}')

print("|Index\t | Nama  \t | Harga\t | Stok\t |")

while menuYangDipilihUser != 5:
    if(menuYangDipilihUser == 1): 
        for key, item in dataTokoBuah.items():
            print(f"|{item['Index']}\t | {key}  \t | {item['Harga']}\t | {item['Stok']}\t |")
        menuYangDipilihUser = int(input('Masukkan angka Menu yang ingin di jalankan: '))
        print(f'Menu yang dipilih adalah : {menuYangDipilihUser}')

    elif(menuYangDipilihUser == 2):
        buahYangInginDitambahkan = input('Masukan Nama Buah : ')
        stokBuahYangDitambahkan = int(input('Masukan Stok Buah : '))
        hargaBuahYangDitambahkan = int(input('Masukan Harga Buah : '))
        dataTokoBuah[buahYangInginDitambahkan] = {
            'Index': 0,
            'Stok': stokBuahYangDitambahkan,
            'Harga': hargaBuahYangDitambahkan
        }
        
        tempList = list(dataTokoBuah.keys())
        for idx, item in enumerate(tempList):
            dataTokoBuah[item]['Index'] = idx
        menuYangDipilihUser = int(input('Masukkan angka Menu yang ingin di jalankan: '))
        print(f'Menu yang dipilih adalah : {menuYangDipilihUser}')

    elif(menuYangDipilihUser == 3):
        buahYangMauDiDelete = int(input('Masuk Index Buah Yang Ingin Di Delete : '))
        
        for key, item in dataTokoBuah.copy().items():
            if item['Index'] == buahYangMauDiDelete:
                del dataTokoBuah[key]

        tempList = list(dataTokoBuah.keys())
        for idx, item in enumerate(tempList):
            dataTokoBuah[item]['Index'] = idx
        menuYangDipilihUser = int(input('Masukkan angka Menu yang ingin di jalankan: '))
        print(f'Menu yang dipilih adalah : {menuYangDipilihUser}')

    elif(menuYangDipilihUser == 4):
        nomorBuahYangInginDibeli = int(input('Masukan Nomor Buah Yang Ingin Dibeli : '))
        stokBuahYangInginDibeli = int(input('Berapa Buah Yang Ingin Anda Beli? '))

        for key, item in dataTokoBuah.items():
            if item['Index'] == nomorBuahYangInginDibeli:
                print(item['Index'], stokBuahYangInginDibeli)
                while stokBuahYangInginDibeli > item['Stok']:
                    print(f'Jumlah Stok Buah Hanya {item['Stok']}, Masukan Stok Yang Lebih Kecil')
                    stokBuahYangInginDibeli = int(input('Berapa Buah Yang Ingin Anda Beli? '))

                kantongBelanjaUser[key] = {
                    'Stok': stokBuahYangInginDibeli,
                    'Harga': item['Harga']
                }

                dataTokoBuah[key]['Stok'] -= stokBuahYangInginDibeli
        
                    

        for key, item in kantongBelanjaUser.items():
            print(f"| {key}  \t | {item['Harga']}\t | {item['Stok']}\t |")
        
        apakahUserMauBeliLagi = input('Apakah anda ingin membeli buah lagi? (Ya/Tidak) ')
        while apakahUserMauBeliLagi.lower() == 'ya':
            nomorBuahYangInginDibeli = int(input('Masukan Nomor Buah Yang Ingin Dibeli : '))
            stokBuahYangInginDibeli = int(input('Berapa Buah Yang Ingin Anda Beli? '))
            
            for key, item in dataTokoBuah.items():
                if item['Index'] == nomorBuahYangInginDibeli:
                    while stokBuahYangInginDibeli > item['Stok']:
                        print(f'Jumlah Stok Buah Hanya {item['Stok']}, Masukan Stok Yang Lebih Kecil')
                        stokBuahYangInginDibeli = int(input('Berapa Buah Yang Ingin Anda Beli? '))

                    if key in kantongBelanjaUser:
                        kantongBelanjaUser[key]['Stok'] += stokBuahYangInginDibeli
                    else:
                        kantongBelanjaUser[key] = {
                            'Stok': stokBuahYangInginDibeli,
                            'Harga': item['Harga']
                        }                        

                    dataTokoBuah[key]['Stok'] -= stokBuahYangInginDibeli
                else:
                    raise TypeError('Nomor yang dimasukkan tidak ada di menu buah')
            apakahUserMauBeliLagi = input('Apakah anda ingin membeli buah lagi? (Ya/Tidak) ')

        
        for key, item in kantongBelanjaUser.items():
            print(f"{key:<15} {item['Stok']:<5} {item['Harga']:<10}")

        totalBelanjaUser = 0
        print('\nDetail Belanja\n')

        for key, item in kantongBelanjaUser.items():
            print(f'{key} : {item['Stok']} x {item['Harga']} = {item['Stok'] * item['Harga']}')
            totalBelanjaUser += item['Stok'] * item['Harga']
        print(f'Total Belanja : Rp{totalBelanjaUser}')
        inputUangUser = int(input('Masuk Jumlah Uang: '))
        print(f'Uang anda sebesar: {inputUangUser}')

        while (inputUangUser - totalBelanjaUser) < 0:
            print(f'Transaksi anda dibatalkan\nuangnya kurang sebesar Rp{abs(inputUangUser - totalBelanjaUser)}')
            inputUangUser = int(input('Masuk Jumlah Uang: '))
        if (inputUangUser - totalBelanjaUser) > 0:
            print(f'Terima Kasih\nUang Kembali anda : Rp{inputUangUser - totalBelanjaUser}')
            break
        else:
            print('Terima Kasih')
            break

print('Terima Kasih Sudah Berbelanja di Toko Buah Aldy!')
```
