```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
X, y, centers = make_blobs(
    n_samples  = 1000,
    n_features = 5,
    shuffle    = True,
    random_state=2020,
    return_centers=True
)
```

Create Clustering Model

```python
sns.scatterplot(
    x=X[:,0],
    y=X[:,1],
    hue=y
)
```

```python
KMeansModel = KMeans(
    n_clusters=3,
    random_state=2020
)
```

```python
KMeansModel.fit(X)
```

```python
y_pred = KMeansModel.predict(X=X)
```

```python
y_pred
```

```python
sns.scatterplot(
    x=X[:,0],
    y=X[:,1],
    hue=y_pred
)
```

Clustering Model Assessment: Elbow Method

```python
num_of_cluster = [x for x in range(1,11)]
model_inertias = []

for clusters in num_of_cluster:
    KMeansModel = KMeans(
        n_clusters=clusters,
        random_state=2020,
        verbose=True
    ).fit(X)

    model_inertias.append(KMeansModel.inertia_)
```

```python
sns.lineplot(
    x = num_of_cluster,
    y = model_inertias
)
sns.scatterplot(
    x = num_of_cluster,
    y = model_inertias
)
plt.xlabel("Jumlah Cluster")
plt.ylabel("Model Inertia")
plt.show()
```

Clustering Assessment: Silhoutte Score

```python
num_of_cluster = [x for x in range(2,11)]
silhouette_scores = []

for clusters in num_of_cluster:
    KMeansModel = KMeans(
        n_clusters=clusters,
        random_state=42,
        verbose=True
    ).fit(X)

    y_pred = KMeansModel.predict(X)
    silhouette_scores.append(silhouette_score(X=X, labels=y_pred, random_state=2020))
```

```python
sns.lineplot(
    x=num_of_cluster,
    y=silhouette_scores
)
sns.scatterplot(
    x = num_of_cluster,
    y = silhouette_scores
)
plt.xlabel("Jumlah Cluster")
plt.ylabel("Silhouette scores")
plt.show()
```
